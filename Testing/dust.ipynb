# Find normalization frequency $\nu_0$

```python
import Models.Dust as Dust
importlib.reload(Dust)

AmodeodustAH = Dust.Amodeo2021({'model':'ACT+Hershel'})
AmodeodustH = Dust.Amodeo2021({'model':'Hershel'})
```

```python
xdatalistAH, xdatalistH = [x500ah, x350ah, x250ah], [x500h, x350h, x250h]
ydatalistAH, ydatalistH = [y500ah, y350ah, y250ah], [y500h, y350h, y250h]
nus = [600, 857, 1200]

def residuals(param_array, nu0):
    res = []
    for nu, x, y in zip(nus, xdatalistAH, ydatalistAH):
        p = {list(AmodeodustAH.params.keys())[i]: param_array[i] for i in range(6)}
        res.append(y - AmodeodustAH.dustpoly(x, nu, nu0=nu0)(p))
    for nu, x, y in zip(nus, xdatalistH, ydatalistH):
        p = {list(AmodeodustH.params.keys())[i]: param_array[i+6] for i in range(6)}
        res.append(y - AmodeodustH.dustpoly(x, nu, nu0=nu0)(p))
    return np.concatenate(res)

result857_I = scipy.optimize.least_squares(lambda param_array: residuals(param_array, nu0=857), list(AmodeodustAH.p0.values())+list(AmodeodustH.p0.values()))
AHdict_I_857 = {list(AmodeodustAH.p0.keys())[i]: np.round(result857_I.x.value[i], 4) for i in range(6)}
Hdict_I_857 = {list(AmodeodustH.p0.keys())[i]: np.round(result857_I.x.value[i+6], 4) for i in range(6)}

result1200_I = scipy.optimize.least_squares(lambda param_array: residuals(param_array, nu0=1200), list(AmodeodustAH.p0.values())+list(AmodeodustH.p0.values()))
AHdict_I_1200 = {list(AmodeodustAH.p0.keys())[i]: np.round(result1200_I.x.value[i], 4) for i in range(6)}
Hdict_I_1200 = {list(AmodeodustH.p0.keys())[i]: np.round(result1200_I.x.value[i+6], 4) for i in range(6)}

result600_I = scipy.optimize.least_squares(lambda param_array: residuals(param_array, nu0=600), list(AmodeodustAH.p0.values())+list(AmodeodustH.p0.values()))
AHdict_I_600 = {list(AmodeodustAH.p0.keys())[i]: np.round(result600_I.x.value[i], 4) for i in range(6)}
Hdict_I_600 = {list(AmodeodustH.p0.keys())[i]: np.round(result600_I.x.value[i+6], 4) for i in range(6)}

print(f"857GHz {AHdict_I_857}, \n1200GHz {AHdict_I_1200}, \n600GHz {AHdict_I_600}")
```
