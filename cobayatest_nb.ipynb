```python
A_dust = 0.2
T_dust = 20*u.K
beta_dust = 1.1
c0 = 6
c1 = -2
c2 = 0.1
nu0 = (c.c/(800*1e-6*u.m)).to(u.GHz)
z = 0.55

def dust(nu, R):
    x = lambda nu: np.exp((nu*c.h/c.k_B/T_dust).decompose()) - 1
    return A_dust*(nu*(1+z)/nu0)**(beta_dust+3) * x(nu0)/x(nu*(1+z)) * (c0+c1*R+c2*R**2) * (u.kJy/u.sr)
```
